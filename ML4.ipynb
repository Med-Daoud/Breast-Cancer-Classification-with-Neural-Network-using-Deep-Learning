```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
data=pd.read_csv('data.csv')
data.tail()
```

```python
from sklearn.preprocessing import OrdinalEncoder,StandardScaler

encoder = OrdinalEncoder()

data[['diagnosis']] = encoder.fit_transform(data[['diagnosis']])
data.tail()
```

```python
data.describe()
```

```python
data.isnull().sum()
```

```python
data=data.drop(data[['id','Unnamed: 32']],axis=1)
data
```

```python
data.isnull().sum()
```

```python
data.groupby('diagnosis').mean()
```

```python
data['diagnosis'].value_counts()
```

```python
y=data['diagnosis']
x=data.drop('diagnosis',axis=1)
x.head()
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)

scaler = StandardScaler()

x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
x_train
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
```

```python
model = keras.Sequential([
    layers.Flatten(input_shape=(30,)),  # Input layer
    layers.Dense(30, activation='relu'),  # Hidden layer
    layers.Dense(2, activation='sigmoid')  # Output layer
])
```

```python
model.compile(
    optimizer = 'adam' ,
    loss='sparse_categorical_crossentropy' ,
    metrics = ['accuracy']
)
```

```python
model.fit(x_train,y_train,epochs=20)
```

```python
loss,accuracy = model.evaluate(x_test,y_test)
```

```python
predic=model.predict(x_test)
np.argmax(predic[0])
```

```python
predictions = [int(np.argmax(i)) for i in predic]
print(predictions)
```
